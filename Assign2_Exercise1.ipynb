```python
#Maryam Assaedi
#Mst. Mahfuja Akter
#Mahpara Hyder Chowdhury


import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.colors as clr
import numpy as np

#a
img = mpimg.imread('oldtimer.png')
plt.imshow(img)
```

### 

```python
#b
hsv_img = clr.rgb_to_hsv(img)
value_img= hsv_img[:,:,2]
plt.imshow(value_img,cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
```

```python
#c
gray_img = np.mean(img, -1)
plt.imshow(gray_img,cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
```

```python
#d
hsv_img[:,:,1] *= 0.5
plt.imshow(hsv_img)
```

```python
#e
image = np.zeros((388,640,3), dtype="uint8")
image[np.where((image==[0,0,0]).all(axis=2))] = [255,222,173]
brown_img=np.divide(image,255)
sat_img_rgb= clr.hsv_to_rgb(hsv_img)
old_img=np.add(brown_img,sat_img_rgb)
rgb_old_img = np.divide(old_img,2)
plt.imshow(rgb_old_img)

```

```python
#f
hsv_diff_hues=clr.rgb_to_hsv(img)
hsv_diff_hues[:,:,0]*=0.6
rgb=clr.hsv_to_rgb(hsv_diff_hues)
plt.imshow(rgb)

```

```python
hsv_diff_hues2 =clr.rgb_to_hsv(img)
hsv_diff_hues2[:,:,0]*=0.8
rgb2=clr.hsv_to_rgb(hsv_diff_hues2)
plt.imshow(rgb2)
```
