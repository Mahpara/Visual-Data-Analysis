```python
"""
Created on Sat May 11 14:28:18 2019

#Mahpara Hyder Chowdhury
#Maryam Assaedi
#Mst. Mahfuja Akter

"""

#a

import pandas as pd  
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot 

init_notebook_mode(connected=True) 

import cufflinks as cf  
cf.go_offline() 



df = pd.read_excel("Data_Cortex_Nuclear.xls")
#df.mean()
df2 = df.fillna(0)


col_names = list(df2) # get column names from existing dataframe
groups = df2.groupby('class') # creates a new object of groups of data
output_df = pd.DataFrame(columns = col_names)


df3 = df2[df2.loc[:, 'class'].isin(df[df['class'] == 'c-SC-s']['class'])]
df3 = df3.append(df2[df2.loc[:, 'class'].isin(df[df['class'] == 't-SC-s']['class'])])



df3['class'].replace(['c-SC-s','t-SC-s'],[1,2], inplace = True)
finaldf = df3[['pPKCG_N', 'pP70S6_N', 'pS6_N', 'pGSK3B_N', 'ARC_N', 'class']]


data = [
    go.Parcoords(
        line = dict(color = finaldf['class']),
        dimensions = list([
            dict(range = [0,3.3],
                label = 'pPKCG_N', values = finaldf['pPKCG_N']),
            dict(range = [0,3.3],
                label = 'pP70S6_N', values = finaldf['pP70S6_N']),
            dict(range = [0,3.3],
                label = 'pS6_N', values = finaldf['pS6_N']),
            dict(range = [0,3.3],
                label = 'pGSK3B_N', values = finaldf['pGSK3B_N']),
            dict(range = [0,3.3],
                label = 'ARC_N', values = finaldf['ARC_N'])
        ])
    )
]

layout = go.Layout(
    plot_bgcolor = '#E5E5E5',
    paper_bgcolor = '#E5E5E5'
)

fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
