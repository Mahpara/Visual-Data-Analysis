```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import forcelayout as fl

correlations = df.corr()
correlations.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
import matplotlib.pyplot as plt
import pydot
import os
import pandas as pd
from IPython.display import Image, display
import graphviz as gv
import tempfile
#import os
#os.environ["PATH"] += os.pathsep + 'C:/Users/jed/Anaconda3/envs/keras/Library/bin/graphviz/'
def edges(matr):
    edges = {}
    nodes=list(matr)
    for m in matr.columns:
        for n in matr.index:
            a,b = m,n 
            if a > b:#only add edge once
                x = matr.at[m, n]
                if(x>0.6):
                    edges[m,n] = float("{0:.4f}".format(x))
    return (edges,nodes)


A =pd.DataFrame(data = correlations )
np.fill_diagonal(A.values, 1)

e,n = edges(A)
g = gv.Graph(engine='fdp',format="png")

for node in n:
    maxVal=0
    for k,v in e.items():
        if (k[0]==node) | (k[1]==node):
            if v>maxVal:
                maxVal=v
    if maxVal>0.9:
        g.node(node,fillcolor='navy', style='filled')
    elif (0.8<maxVal) and (maxVal<=0.9):
        g.node(node,fillcolor='blue', style='filled')
    elif (0.6<maxVal) and (maxVal<=0.8):
        g.node(node,fillcolor='lightblue', style='filled')
    elif maxVal<=0.6:
        g.node(node,fillcolor='white', style='filled')

for k, v in e.items():
    if(v > 0.6):
        if v<0.7:
            g.edge(k[0], k[1], len=str(v),penwidth=str(1))
        elif v<0.8:
            g.edge(k[0], k[1], len=str(v),penwidth=str(2))
        elif v<0.9:
            g.edge(k[0], k[1], len=str(v),penwidth=str(3))
        elif v>0.9:
            g.edge(k[0], k[1], len=str(v),penwidth=str(4))
        
    
g.render('graph', view = True)
```
